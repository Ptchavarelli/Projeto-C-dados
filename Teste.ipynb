# Funcao antiga de limpeza nao usar mais

```python
###Fun√ßao Limpar N√£o Usar mais

def Limpar(frases):
    strings=[]
    for i in frases:
        i.replace('/n','')
        i.replace('!','')
        i.replace("'","")
        junto= i.split(" ")
        for w in junto:
            if '@' in w:
                del w
            elif 'http' in w:
                del w
            elif w == 'rt':
                  del w
            elif '\n' in w:
                del w
            else:
                x = list(w)
                for p in x:
                    if p == '?':
                        x.remove(p)
                    elif p == ',':
                        x.remove(p)
                    elif p == '.':
                        x.remove(p)
                    elif p == '!':
                        x.remove(p)
                    elif p == '#':
                        x.remove(p)
                    elif p == '&':
                        x.remove(p)
                    elif p == ';':
                        x.remove(p)
                    elif p == '(':
                        x.remove(p)
                    elif p == ')':
                        x.remove(p)
                    elif p == '"':
                        x.remove(p)
                    elif p == '/':
                        x.remove(p)
                    elif p == '*':
                        x.remove(p)
                    elif p == '-':
                        x.remove(p)
                tudo= ''.join(x)
                strings.append(tudo)
    return strings
```

```python
def Palavra_relevante(palavra):
    soma=1
    total=0
    for i in Palavras_relevantes:
        total+=len(i)
        for e in i:
            if e.lower()==palavra.lower():
                soma+=1
    return soma/(total+len(Laplace))
            
    
    
    
#print(Palavra_relevante('corolla'))
#print(Palavra_relevante('oi'))
#print(Palavra_relevante('namorada'))
#print(Palavra_relevante('ruim'))
```

# Probabilidade de uma palavra ser irrelevante

```python
def Palavra_irrelevante(palavra):
    soma=1
    total=0
    for i in Palavras_irrelevantes:
        total+=len(i)
        for e in i:
            if e.lower()==palavra.lower():
                soma+=1
    return soma/(total+len(Laplace))
            
    
    
    
#print(Palavra_irrelevante('corolla'))
#print(Palavra_irrelevante('oi'))
#print(Palavra_irrelevante('namorada'))
#print(Palavra_irrelevante('ruim'))
```

# Classifica√ß√£o

```python
def Classifica√ß√£o(lista):
    classificado = []
    for i in lista:         
        palavra = i.split()
        Prelevante= []
        Pirrelevante = []
        for e in palavra:         
            Prelevante.append(Palavra_relevante(e))
            Pirrelevante.append(Palavra_irrelevante(e))
        PR = 1
        PI = 1
        for i in Prelevante:
            PR *= i
        for i in Pirrelevante:
            PI *= i
        PRfrase = PR * Pr 
        PIfrase = PI * Pirr
        if PRfrase > PIfrase:
            classificado.append('Relevante')
        else:
            classificado.append('Irrelevante')
    return classificado
```
