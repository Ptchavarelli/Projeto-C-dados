```python
##Funcao que limpa


def Limpar(frases):
    strings=[]
    for i in frases:
        i.replace('/n','')
        i.replace('!','')
        i.replace("'","")
        junto= i.split(" ")
        for w in junto:
            if '@' in w:
                del w
            elif 'http' in w:
                del w
            elif w == 'rt':
                  del w
            elif '\n' in w:
                del w
            else:
                x = list(w)
                for p in x:
                    if p == '?':
                        x.remove(p)
                    elif p == ',':
                        x.remove(p)
                    elif p == '.':
                        x.remove(p)
                    elif p == '!':
                        x.remove(p)
                    elif p == '#':
                        x.remove(p)
                    elif p == '&':
                        x.remove(p)
                    elif p == ';':
                        x.remove(p)
                    elif p == '(':
                        x.remove(p)
                    elif p == ')':
                        x.remove(p)
                    elif p == '"':
                        x.remove(p)
                    elif p == '/':
                        x.remove(p)
                    elif p == '*':
                        x.remove(p)
                    elif p == '-':
                        x.remove(p)
                tudo= ''.join(x)
                strings.append(tudo)
    return strings
```
